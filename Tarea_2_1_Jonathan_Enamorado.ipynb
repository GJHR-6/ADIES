```python
import pandas as pd
import seaborn as sb

datos = pd.read_csv('U4_01_housing.csv')
```

```python
datos.head()
```

```python
datos_na = datos.dropna()
datos_na.head()
```

```python
dummies = pd.get_dummies(datos_na['ocean_proximity'], dtype=int)
datos_na = datos_na.join(dummies)
datos_na = datos_na.drop(['ocean_proximity'], axis=1)
```

```python
datos_na.hist( figsize=(15,8), bins=30, edgecolor = 'black' )
```

```python
datos_na = datos_na[(datos_na['housing_median_age'] < 50)]
```

```python
datos_na.corr()
```

```python
sb.set(rc={'figure.figsize':(15,8)})
sb.heatmap(data=datos_na.corr(), annot=True, cmap='YlGnBu' )
```

```python
datos.corr()['housing_median_age'].sort_values(ascending=False)
```

```python
sb.scatterplot(x=datos_na['median_house_value'], y=datos_na['housing_median_age'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = datos_na.drop('median_house_value', axis=1)
y = datos_na[['median_house_value']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=831)
```

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)
modelo.score(X_test, y_test)
```
